```python
import time
def fetch_vendor_data(vendor_ids):
    mock_db = {
                "V001": {"name": "Alpha Corp", "rating": 4.5},
                "V002": {"name": "Beta Ltd", "rating": 3.8},
                "V004": {"name": "Delta Inc", "rating": 4.9},
                }
    results = {}
    for vendor_id in vendor_ids:
        time.sleep(0.5)
        try:
            results[vendor_id] = mock_db[vendor_id]
        except KeyError:
            print(f"Vendor ID '{vendor_id}' not found in database.")  
    return results
vendor_list = ["V001", "V002", "V003", "V004"]
data = fetch_vendor_data(vendor_list)
print("\nFinal Result:")
print(data)
```
